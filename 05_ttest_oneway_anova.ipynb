# Bivariate:  Cat/Num

* **student's t-test** allows you to compare two *groups*
* **one-way analysis of variance** which allows you to compare three or more *groups*


*group* is a term to refer to the category values in a particular feature.




## 01. Student's t-test (t) 

Its purpose is to determine whether there is a statistically significant difference between the means of a numeric variable for two values of categorical variable.


### 01.01. Histogram visualization

**Mean**

If we can see the two means far enough in each one of the distributions the we conclude that the categorical feature has statistically significant relationship (important effect size) on the numeric variable.


**Std**

We need to review how spread out each one of the distribution is. If there is so much crossover between the two distributions (so much variance) is difficult to know if the means are really different. 

**Means apart & Small std:**
* big t-value
* low p-value
* large effect size or relationship

**Means apart & High std:**
* normal t-value
* normal p-value
* smaller but still significant relationship

**Means close & High std:**
* small t-value
* high p-value
* no effect or relationship

### 01.02. Formula

$t = \frac{(x_1 - x_2)}{\sqrt{\frac{(s_1)^2}{n_1} + \frac{(s_2)^2}{n_2}}}$


$x_1,x_2$  means of the groups

$s_1,s_2$ standard deviations

$n_1,n_2$ sample sizes (for normalize the group sizes)

The t-value gets higher if the means difference is high or the variance is low 

### 01.03. p-value

To calculate the associated p-value is generated by calculating a confidence interval and the DoF (degrees of freedom)


### 01.04 One-Tailed test

If we have a theorical reason to justify that one group in particulas will be higher or lower that the other, the we want to have a p-value based on *one-sided* criterion

## 02. One-way ANOVA (F)

Collection of statistical models used to analyze the differences among group means in a sample. It compares the differences in numeric values across a set of categorical values

### 02.01. Formula

$k$: number of groups

$\bar{x}:$ mean of the numeric variable

$\bar{x}_j$: mean of the numeric variable selecting the group $j$


####   between-group mean square value

$SSB = \sum_jn_j(\bar{x}_j - \bar{x})^2$

$df_b = k - 1$

$MSB = \frac{SSB}{df_b}$

#### within-group mean square value

(variance of each group) 

$SSW = \sum_j \sum_i (\bar{x}_{ij} - \bar{x}_j)^2$

$df_b = n - k$

#### F-ratio

$F = \frac{MSB}{MSW}$

## 03. Assumptions

* Normal Distribution of the numeric variable - for each group of the categorical variable

* Equal variance of the numeric variable for each group of the categorical variable (if the sample size across the groups are close to equals then this assumption is less important)

* The data used to carry out the analysis were sampled independently between the groups
